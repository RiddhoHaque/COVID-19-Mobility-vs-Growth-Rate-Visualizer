```python
import pandas as pd
import pandasql as ps

countries=['Italy', 'Bangladesh', 'Vietnam', 'Sweden', 'New Zealand', 'Singapore', 'Indonesia', 'Spain', 'Portugal', 'UK', 'Japan', 'Sri Lanka']
colours=['tab:brown', 'tab:green', 'tab:orange', 'b', 'k', 'tab:pink', 'm', 'y', 'tab:purple', 'tab:cyan', 'tab:red', 'tab:brown']
pop_density=[206, 1115.62, 314, 25, 46, 8358, 140.08, 243, 288, 430, 899, 341]
population=[60461826, 164689383, 97338579, 10099265, 4822233, 5850342, 273397681, 46754118,  10197799, 67871466, 126490626, 21412817]
confirmed_df=pd.read_csv('time_series_covid19_confirmed_global.csv')
recovered_df=pd.read_csv('time_series_covid19_recovered_global.csv')
death_df=pd.read_csv('time_series_covid19_deaths_global.csv')

def get_day(days, months):
    if(months>=2):
        days+=31
    if(months>=3):
        days+=29
    if(months>=4):
        days+=31
    if(months>=5):
        days+=30
    if(months>=6):
        days+=31
    if(months>=7):
        days+=30
    if(months>=8):
        days+=31
    if(months>=9):
        days+=31
    if(months>=10):
        days+=30
    if(months>=11):
        days+=31
    if(months>=12):
        days+=30
    return days

def get_day_csse(str):
    parts=str.split("/")
    months=int(parts[0])
    days=int(parts[1])
    years=int(parts[2])
    return get_day(days, months)

def get_day_mobility(str):
    parts=str.split("-")
    years=int(parts[0])
    months=int(parts[1])
    days=int(parts[2])
    return get_day(days, months)

st_date=get_day_mobility("2020-02-15")
ed_date=get_day_mobility("2020-06-27")
print(get_day_csse("6/1/20"))

case_info=[[0 for j in range(len(countries))] for i in range(366)]
mobility_info=[[0 for j in range(len(countries))] for i in range(366)]
countryIndex=0
for country in countries:
    country_alt=country
    if(country=='UK'):
        country_alt='United Kingdom';
    qCon="Select * from confirmed_df where [Country/Region]='"+country_alt+"' and [Province/State] is null"
    qRec="Select * from recovered_df where [Country/Region]='"+country_alt+"' and [Province/State] is null"
    qDeath="Select * from death_df where [Country/Region]='"+country_alt+"' and [Province/State] is null"
    cases_country=ps.sqldf(qCon, locals())
    recovered_country=ps.sqldf(qRec, locals())
    death_country=ps.sqldf(qDeath, locals())
    for column in cases_country.columns:
        if(column[0].isdigit()):
            day=get_day_csse(column)
            val=0
            for idx, row in cases_country.iterrows():
                val=row[column]
            if(day>=st_date and day<=ed_date):
                if countryIndex==0 and day==get_day_csse("6/1/20"):
                    print('Confirmed '+str(val))
                case_info[day][countryIndex]=val
    for column in recovered_country.columns:
        if(column[0].isdigit()):
            day=get_day_csse(column)
            val=0
            for idx, row in recovered_country.iterrows():
                val=row[column]
            if(day>=st_date and day<=ed_date):
                if countryIndex==0 and day==get_day_csse("6/1/20"):
                    print('Recovered '+str(val))
                case_info[day][countryIndex]-=val
    for column in death_country.columns:
        if(column[0].isdigit()):
            day=get_day_csse(column)
            val=0
            for idx, row in death_country.iterrows():
                val=row[column]
            if(day>=st_date and day<=ed_date):
                if countryIndex==0 and day==get_day_csse("6/1/20"):
                    print('Deaths '+str(val))
                case_info[day][countryIndex]-=val
    qMob="Select * from df where country_region='"+country_alt+"' and sub_region_1 is NULL"
    mob_df=ps.sqldf(qMob, locals())
    for idx, rows in mob_df.iterrows():
        day=get_day_mobility(rows["date"])
        val=rows["retail_and_recreation_percent_change_from_baseline"]
        val+=rows["transit_stations_percent_change_from_baseline"]
        val+=rows["workplaces_percent_change_from_baseline"]
        val/=3
        mobility_info[day][countryIndex]=val
    countryIndex+=1 

rev_mobility_info=[[0 for j in range(len(countries))] for i in range(366)]

running_avg_interval=3
for day in range(366):
    for countryIndex in range(len(countries)):
        sum=0
        cnt=0
        for i in range(running_avg_interval):
            if day-i>=st_date:
                sum+=mobility_info[day-i][countryIndex]
                cnt+=1
        if(cnt==0):
            rev_mobility_info[day][countryIndex]=sum
        else:
            rev_mobility_info[day][countryIndex]=(sum*1.0)/(cnt*1.0)
mobility_info=[[rev_mobility_info[i][j] for j in range(len(countries))] for i in range(366)]
```

```python
#from matplotlib import rcParams
import math
import matplotlib.pyplot as plt
from matplotlib import rcParams
#rcParams.update({'figure.autolayout': True})
def get_date(day):
    days_in_month=[31, 29, 31, 30, 31, 30, 31]
    idx=0
    month=1
    while days_in_month[idx]<day:
        month+=1
        day-=days_in_month[idx]
        idx+=1
    return str(day)+"/"+str(month)

def distort(val):
    if val==0:
        return 0
    #print(val)
    return math.log10(val)
#print(distort(100000))

def day_num(day):
    if day<10:
        return '00'+str(day)
    if day<100:
        return '0'+str(day)
    return str(day)

mx=5
for i in range(365):
    for j in range(len(countries)):
        if(case_info[i][j]<0):
            print(countries[j], get_date(i), case_info[i][j])
        if(case_info[i][j]>mx):
            mx=distort(case_info[i][j])
for i in range(365):
    if i>=st_date and i<=ed_date:
        active_cases=[distort(case_info[i][j]) for j in range(len(countries))]
        #for j in range(len(population)):
        #    active_cases_per_million[j]/=population[j]
        #    active_cases_per_million[j]*=10000000
        x=active_cases
        #ticks = [0, 5000, 20000, 50000, 100000]
        #plt.xticks(range(5), ticks)
        y=mobility_info[i]
        c=colours
        s=pop_density
        fig = plt.figure(figsize=(8, 8))
        plt.ylim([-100, 30])
        plt.xlim([0, 6])
        ticks=[0, 10, 100, 1000, 10000, 100000, 1000000]
        plt.xticks(range(7), ticks)
        plt.scatter(x=x, y=y, s=s, alpha=0.5, c=c, cmap="Blues", edgecolors="grey", linewidth=2)
        plt.title('Community mobility changes vs COVID19 cases-'+get_date(i))
        plt.ylabel('Mobility change (%)')
        plt.xlabel('Active cases')
        #plt.scatter(x=x, y=y, s=s, alpha=0.5, c=c)
        idx=0
        for country in countries:
            plt.annotate(country, (x[idx], y[idx]))
            idx+=1
        filename='step'+day_num(i)+'.png'
        plt.savefig(filename, dpi=96)
        plt.gca()
```

```python
import pandas as pd
import pandasql as ps
import math
import matplotlib.pyplot as plt

mobility_df=pd.read_csv('Global_Mobility_Report.csv')
confirmed_df=pd.read_csv('time_series_covid19_confirmed_global.csv')
recovered_df=pd.read_csv('time_series_covid19_recovered_global.csv')
death_df=pd.read_csv('time_series_covid19_deaths_global.csv')

qCountries="Select distinct country_region from mobility_df"
countries=ps.sqldf(qCountries, locals())
def get_day(days, months):
    if(months>=2):
        days+=31
    if(months>=3):
        days+=29
    if(months>=4):
        days+=31
    if(months>=5):
        days+=30
    if(months>=6):
        days+=31
    return days

def get_day_csse(str):
    parts=str.split("/")
    months=int(parts[0])
    days=int(parts[1])
    years=int(parts[2])
    return get_day(days, months)

def get_day_mobility(str):
    parts=str.split("-")
    years=int(parts[0])
    months=int(parts[1])
    days=int(parts[2])
    return get_day(days, months)


st_date=get_day_mobility("2020-02-15")
ed_date=get_day_csse("5/29/20")
ed_date_csse=get_day_csse("6/5/20")

case_info=[[0 for j in range(len(countries))] for i in range(366)]
case_available=[[False]*len(countries) for i in range(366)]
mobility_info=[[0 for j in range(len(countries))] for i in range(366)]
mobility_available=[[False for j in range(len(countries))] for i in range(366)]
countryIndex=0
for country in countries["country_region"]:
    country_alt=country
    if country == "The Bahamas":
        country="Bahamas"
    if country == "CÃ´te d'Ivoire":
        country="Cote d Ivoire"
    if country == "Cape Verde":
        country="Cabo Verde"
    if country=="South Korea":
        country="Korea, South"
    if country=="Myanmar (Burma)":
        country="Burma"
    if country=="Taiwan":
        country="Taiwan*"
    if country=="United States":
        country="US"
    qCon=""
    qRec=""
    qDeath=""
    if country=="Hong Kong":
        qCon="Select * from confirmed_df where [Country/Region]='China' and [Province/State]='"+country+"'"
        qRec="Select * from recovered_df where [Country/Region]='China' and [Province/State]='"+country+"'"
        qDeath="Select * from death_df where [Country/Region]='China' and [Province/State]='"+country+"'"
    else:
        qCon="Select * from confirmed_df where [Country/Region]='"+country+"'"
        qRec="Select * from recovered_df where [Country/Region]='"+country+"'"
        qDeath="Select * from death_df where [Country/Region]='"+country+"'"
    cases_country=ps.sqldf(qCon, locals())
    recovered_country=ps.sqldf(qRec, locals())
    death_country=ps.sqldf(qDeath, locals())
    for column in cases_country.columns:
        if(column[0].isdigit()):
            day=get_day_csse(column)
            val=0
            for idx, row in cases_country.iterrows():
                val+=row[column]
            if(day>=st_date and day<=ed_date):
                case_info[day][countryIndex]=val
                case_available[day][countryIndex]=True
    for column in recovered_country.columns:
        if(column[0].isdigit()):
            day=get_day_csse(column)
            val=0
            for idx, row in recovered_country.iterrows():
                val+=row[column]
            if(day>=st_date and day<=ed_date):
                case_info[day][countryIndex]-=val
    for column in death_country.columns:
        if(column[0].isdigit()):
            day=get_day_csse(column)
            val=0
            for idx, row in death_country.iterrows():
                val+=row[column]
            if(day>=st_date and day<=ed_date):
                case_info[day][countryIndex]-=val
    qMob="Select * from mobility_df where country_region='"+country_alt+"' and sub_region_1 is NULL"
    mob_df=ps.sqldf(qMob, locals())
    for idx, rows in mob_df.iterrows():
        day=get_day_mobility(rows["date"])
        cnt=0
        if(rows["retail_and_recreation_percent_change_from_baseline"] is not None):
            val=rows["retail_and_recreation_percent_change_from_baseline"]
            cnt+=1
        if(rows["transit_stations_percent_change_from_baseline"] is not None):
            val+=rows["transit_stations_percent_change_from_baseline"]
            cnt+=1
        if(rows["workplaces_percent_change_from_baseline"] is not None):
            val+=rows["workplaces_percent_change_from_baseline"]
            cnt+=1
        val/=cnt
        if cnt==0:
            continue
        mobility_info[day][countryIndex]=val
        mobility_available[day][countryIndex]=True
    countryIndex+=1 

rev_mobility_info=[[0 for j in range(len(countries))] for i in range(366)]

running_avg_interval=3
for day in range(366):
    for countryIndex in range(len(countries)):
        sum=0
        cnt=0
        for i in range(running_avg_interval):
            if day-i>=st_date:
                sum+=mobility_info[day-i][countryIndex]
                cnt+=1
        if(cnt==0):
            rev_mobility_info[day][countryIndex]=sum
        else:
            rev_mobility_info[day][countryIndex]=(sum*1.0)/(cnt*1.0)
mobility_info=[[rev_mobility_info[i][j] for j in range(len(countries))] for i in range(366)]

                
```

```python
def get_doubling_time(st, ed, duration):
    #print(st, ed, duration)
    
    if st==ed:
        return 0
    if ed==1:
        return 0
    if st==0:
        return (duration*math.log(2))/(math.log(ed)) 
    return (duration*math.log(2))/(math.log(ed)-math.log(st))

x=[]
y=[]
c=[]
for day in range(st_date, ed_date+1):
    for ind in range(countryIndex):
        if(mobility_available[day+13][ind] and case_available[day+27][ind]):
            avg_mob=0
            for d in range(14):
                avg_mob+=mobility_info[day+d][ind]
            avg_mob/=14
            if case_info[day+27][ind]>=1000 and case_info[day+14][ind]>=1000 and case_info[day+14][ind]<case_info[day+27][ind]:
                if get_doubling_time(case_info[day+14][ind], case_info[day+27][ind], 14)!=0:
                    x.append(-avg_mob)
                    y.append(1.0/get_doubling_time(case_info[day+14][ind], case_info[day+27][ind], 14))
                    c.append('b')
print(len(x))
print(len(y))
plt.xlabel("Percentage reduction in mobility")
plt.ylabel("Doubling rate of active COVID19 cases")
plt.title("Effect of Lockdown on the doubling rate of COVID-19 worldwide")
plt.scatter(x=x, y=y)
plt.savefig("Lockdown and Doubling rate/Effect of lockdown on doubling rate.png", dpi=100)
plt.gca()
```
